La **summary generation** (ou génération de résumés) est une tâche de traitement automatique du langage naturel (NLP) qui consiste à produire un résumé concis d'un texte tout en conservant les informations essentielles. L'objectif est de réduire le volume du texte tout en préservant la clarté et la pertinence des idées principales.

**Types de Summary Generation**

1. **Résumé Extractif**
   - Cette méthode consiste à extraire les phrases ou segments de texte jugés les plus pertinents directement du texte original.
   - Elle est simple à mettre en œuvre mais peut parfois manquer de fluidité, car elle ne reformule pas le contenu.
   - Exemples : Algorithmes comme TextRank ou des modèles de machine learning basés sur des caractéristiques linguistiques.

2. **Résumé Abstrait**
   - Cette approche génère de nouvelles phrases qui reformulent les idées clés du texte source.
   - Elle est plus complexe mais produit souvent des résumés plus naturels et cohérents.
   - Exemples : Modèles seq2seq avec mécanismes d'attention, transformers comme T5, BERTSUM, ou Pegasus.

**Applications de la Summary Generation**
La génération de résumés est utilisée dans plusieurs domaines :
   - **Journalisme** : Générer des résumés de nouvelles.
   - **Éducation** : Aider les étudiants à réviser des documents longs.
   - **Recherche** : Créer des résumés d’articles académiques.
   - **Service client** : Résumer les conversations ou tickets clients pour un traitement plus rapide.

**Défis de la Summary Generation**
La génération de résumés pose des défis techniques, notamment :
   - **Fidélité** : Garantir que les résumés restent fidèles aux informations du texte source.
   - **Cohérence** : Produire des résumés fluides et lisibles.
   - **Gestion des biais** : Éviter de refléter les biais présents dans les données d'entraînement.

En somme, la summary generation est un domaine essentiel du NLP qui vise à faciliter l'accès rapide à l'information en produisant des résumés automatisés précis et utiles.

**Modèle T5 (Text-To-Text Transfer Transformer)**

**T5** est un modèle de Google conçu pour traiter toutes les tâches NLP dans un format unique de "text-to-text". Cela signifie que, que ce soit pour le résumé, la traduction, ou la classification, toutes les tâches sont reformulées en une tâche de génération de texte, ce qui rend le modèle extrêmement flexible.


#### Ce qui rend T5 spécial :
- **Approche Text-to-Text** : Contrairement à d'autres modèles, T5 utilise un format universel dans lequel les entrées et les sorties sont traitées comme du texte. Par exemple, pour une tâche de résumé, l’entrée pourrait être "résume : [texte]" et la sortie un résumé généré, facilitant l’adaptation à de multiples tâches NLP.
- **Pré-entraînement multitâche** : T5 est pré-entraîné sur une grande variété de tâches en utilisant un corpus appelé C4 (Colossal Clean Crawled Corpus), ce qui le rend particulièrement polyvalent et robuste pour plusieurs applications.

**Architecture de T5**
- **Architecture Encoder-Decoder** : T5 suit une architecture transformer classique de type encoder-decoder, ce qui le différencie des modèles comme BERT (qui est uniquement un encodeur) et GPT (qui est uniquement un décodeur).
- **Masquage de texte** : Pour l’entraînement, T5 masque aléatoirement des tokens dans l'entrée et demande au décodeur de les reconstituer, ce qui améliore ses capacités à prédire et générer des informations manquantes.

- **Efficacité et Performance** : En raison de son architecture et de son approche unifiée, T5 est capable de produire des résumés cohérents et contextuels, rivalisant avec les performances des autres modèles dans de nombreuses tâches NLP, y compris le résumé abtractif.

En résumé, l’approche "text-to-text" et l’architecture encoder-decoder de T5 le rendent unique et très performant pour la génération de résumés, en particulier pour des applications nécessitant de la flexibilité entre différentes tâches NLP.


**Modèle BART (Bidirectional and Auto-Regressive Transformers)**

**BART** est un modèle développé par Facebook, conçu pour exceller dans les tâches de génération de texte comme le résumé, la traduction, et la reformulation. Il combine les forces des modèles bidirectionnels (comme BERT) et autoregressifs (comme GPT), offrant une architecture de type **encoder-decoder**.

#### Ce qui rend BART spécial :
- **Combinaison Encoder-Decoder** : BART utilise un encodeur bidirectionnel qui lit l'ensemble du contexte d'une phrase, combiné avec un décodeur autoregressif qui génère le texte mot par mot. Cette combinaison permet au modèle de bien comprendre le contexte global tout en générant des sorties cohérentes.
- **Pré-entraînement avec perturbation** : BART est pré-entraîné en introduisant des perturbations dans le texte (masquage, permutation de phrases, suppression de mots) et en lui demandant de reconstruire le texte original, ce qui le rend très robuste pour capturer des relations contextuelles profondes.

**Architecture de BART**
- **Encodeur bidirectionnel et décodeur autoregressif** : Contrairement aux modèles comme GPT (qui utilise uniquement un décodeur autoregressif) ou BERT (qui utilise uniquement un encodeur bidirectionnel), BART exploite les deux, permettant une compréhension et une génération de texte particulièrement fines.

- **Flexible pour plusieurs tâches** : Grâce à cette architecture mixte, BART peut être utilisé efficacement pour des tâches extractives et abstraites. Il est particulièrement performant pour les tâches de génération de résumés, car il capture le contexte avec l’encodeur et génère des réponses précises avec le décodeur.

En résumé, l'architecture hybride de BART, associée à son pré-entraînement robuste, le rend unique et particulièrement efficace pour les tâches de génération de texte comme le résumé automatique.

```python
# @title
from transformers import BartForConditionalGeneration,BartTokenizer


bart_tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
bart_model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

def summarize_with_bart(text):
    inputs = bart_tokenizer(text, max_length=1024, truncation=True, return_tensors="pt")
    summary_ids = bart_model.generate(inputs["input_ids"], max_length=128, num_beams=4, length_penalty=2.0, early_stopping=True)
    return bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

```

# **Évaluation avec la Métrique ROUGE**

**La métrique ROUGE est couramment utilisée pour évaluer les systèmes de génération de texte, en particulier dans le domaine du résumé automatique. Elle compare les résumés générés par les modèles aux résumés de référence en calculant des scores basés sur le rappel, la précision et la F1.**


**Types de ROUGE :**

Pour évaluer la qualité des résumés générés par les modèles T5 et BART, nous allons utiliser la métrique ROUGE, qui comprend plusieurs variantes telles que :
    - ROUGE-1 : mesure le chevauchement des mots individuels (unigrammes) entre le résumé généré et le résumé de référence.
    - ROUGE-2 : mesure le chevauchement de paires de mots consécutifs (bigrams) entre le résumé généré et le résumé de référence.
    * ROUGE-L : mesure la plus longue sous-séquence commune (LCS) entre le résumé généré et le résumé de référence.

**Analyse des Scores ROUGE**

Les scores ROUGE obtenus fourniront des indications sur :

  - La qualité du rappel : mesure dans quelle mesure les mots du résumé de référence sont présents dans le résumé généré.
 -  La précision : indique combien des mots présents dans le résumé généré sont également dans le résumé de référence.
  - Le score F1 : une combinaison de la précision et du rappel, donnant une mesure équilibrée de la performance.


```python
# @title
from rouge_score import rouge_scorer

def compute_rouge(reference, generated):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, generated)
    return scores
```

# **Fine-Tuning du Modèle BART**

**Prompt Engineering**

Le **Prompt Engineering** consiste à concevoir des "prompts" ou des instructions données à un modèle pour influencer sa réponse. Pour les modèles de langage, un prompt bien formulé peut guider le modèle vers des réponses plus précises et adaptées.

In-Context Learning (ICL)
**texte en gras**
Dans le cadre d’**ICL**, on donne au modèle des exemples dans le prompt lui-même pour qu’il comprenne mieux ce qu’on attend de lui, sans nécessiter de nouveaux paramètres d’entraînement.

Types d'In-Context Learning :

- **Zero-shot** : Le modèle répond sans aucun exemple contextuel ; on lui pose directement la question.
- **One-shot** : Le modèle reçoit un exemple unique pour illustrer la tâche.
- **Few-shot** : On donne plusieurs exemples pour fournir davantage de contexte et d’indications sur la tâche attendue.

```python
# @title
import time
import psutil
from transformers import BartForConditionalGeneration, BartTokenizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialiser le modèle et le tokenizer BART
bart_tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
bart_model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

def summarize_with_bart(prompt):
    inputs = bart_tokenizer(prompt, max_length=1024, truncation=True, return_tensors="pt")
    summary_ids = bart_model.generate(inputs["input_ids"], max_length=128, num_beams=4, length_penalty=2.0, early_stopping=True)
    return bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

from rouge_score import rouge_scorer

def measure_metrics(prompt, reference_summary):
    # Mesurer le temps de calcul
    start_time = time.time()

    # Mesurer l'utilisation de la mémoire avant la génération
    process = psutil.Process()
    mem_before = process.memory_info().rss

    # Générer le résumé
    generated_summary = summarize_with_bart(prompt)

    # Mesurer l'utilisation de la mémoire après la génération
    mem_after = process.memory_info().rss
    end_time = time.time()

    # Calculer le temps et la mémoire utilisés
    time_taken = end_time - start_time
    memory_used = (mem_after - mem_before) / (1024 * 1024)  # en Mo

    # Calculer la similarité entre le résumé généré et la référence (Cosine Similarity)
    vectorizer = TfidfVectorizer().fit_transform([generated_summary, reference_summary])
    cosine_sim = cosine_similarity(vectorizer[0:1], vectorizer[1:2]).flatten()[0]

    # Calculer le nombre de paramètres du modèle
    num_parameters = bart_model.num_parameters()

    # Calculer les scores ROUGE
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    rouge_scores = scorer.score(reference_summary, generated_summary)

    return {
        "generated_summary": generated_summary,
        "time_taken": time_taken,
        "memory_used": memory_used,
        "similarity_score": cosine_sim,
        "num_parameters": num_parameters,
        "rouge1": rouge_scores['rouge1'].fmeasure,
        "rouge2": rouge_scores['rouge2'].fmeasure,
        "rougeL": rouge_scores['rougeL'].fmeasure
    }


def generate_zero_shot_prompt(article):
    return f"Please summarize the following text:\n\n{article}"

def generate_one_shot_prompt(dataset, idx):
    # Exemples tirés du dataset
    example_article = dataset["test"][idx]["article"]
    example_summary = dataset["test"][idx]["highlights"]

    return (
        f"Example:\nText: {example_article}\nSummary: {example_summary}\n\n"
        f"Please summarize the following text:\nText: {dataset['test'][idx]['article']}"
    )

def generate_few_shot_prompt(dataset, idx):
    # Créer une liste d'exemples tirés du dataset
    few_shot_prompts = ""
    for i in range(1,5):  # Prenons les 3 précédents exemples
        if i >= 0:  # On s'assure que l'indice est valide
            article = dataset["test"][i]["article"]
            summary = dataset["test"][i]["highlights"]
            few_shot_prompts += f"Text: {article}\nSummary: {summary}\n\n"

    # Ajouter l'exemple actuel
    few_shot_prompts += f"Please summarize the following text:\nText: {dataset['test'][idx]['article']}"

    return few_shot_prompts

# Sélectionner un exemple d'article du dataset
idx = 1  # Choisissez l'index d'un article du dataset (par exemple, 1 pour le premier exemple)

# Générer les prompts pour Zero-Shot, One-Shot, et Few-Shot
zero_shot_prompt = generate_zero_shot_prompt(dataset["test"][idx]["article"])
one_shot_prompt = generate_one_shot_prompt(dataset, idx)
few_shot_prompt = generate_few_shot_prompt(dataset, idx)
reference_summary = dataset["test"][idx]["highlights"]

# Mesurer les métriques pour Zero-Shot
zero_shot_metrics = measure_metrics(zero_shot_prompt, reference_summary)
print("Zero-Shot Metrics:", zero_shot_metrics)

# Mesurer les métriques pour One-Shot
one_shot_metrics = measure_metrics(one_shot_prompt, reference_summary)
print("One-Shot Metrics:", one_shot_metrics)

# Mesurer les métriques pour Few-Shot
few_shot_metrics = measure_metrics(few_shot_prompt, reference_summary)
print("Few-Shot Metrics:", few_shot_metrics)
```

```python
# @title
import matplotlib.pyplot as plt

# Données des résultats des différentes méthodes (Zero-Shot, One-Shot, Few-Shot)
methods = ['Zero-Shot', 'One-Shot', 'Few-Shot']
rouge1_scores = [0.494, 0.537, 0.547]
rouge2_scores = [0.253, 0.377, 0.409]
rougeL_scores = [0.427, 0.5, 0.526]
similarity_scores = [0.299, 0.409, 0.378]
time_taken = [17.03, 25.35, 22.96]  # Temps en secondes
memory_used = [-1.0, 0.0, 0.0]  # Mémoire utilisée en Mo

# Création de la figure et des sous-graphes pour les scores ROUGE
fig, ax1 = plt.subplots(figsize=(10, 6))

# Comparaison des scores ROUGE
ax1.bar(methods, rouge1_scores, width=0.2, label='ROUGE-1', align='center', alpha=0.7)
ax1.bar(methods, rouge2_scores, width=0.2, label='ROUGE-2', align='edge', alpha=0.7)
ax1.bar(methods, rougeL_scores, width=0.2, label='ROUGE-L', align='edge', alpha=0.7)

ax1.set_xlabel('Méthode')
ax1.set_ylabel('Scores ROUGE')
ax1.set_title('Comparaison des Résultats pour Zero-Shot, One-Shot, et Few-Shot')
ax1.legend(loc='upper left')

# Ajout des scores de Similarité en utilisant un axe secondaire
ax2 = ax1.twinx()
ax2.plot(methods, similarity_scores, color='black', marker='o', label='Similarité', linestyle='--')
ax2.set_ylabel('Similarité')

# Ajouter les informations de Temps et Mémoire
fig2, ax3 = plt.subplots(figsize=(10, 6))

# Temps d'exécution
ax3.plot(methods, time_taken, color='blue', marker='o', label='Temps (s)', linestyle='--')
ax3.set_xlabel('Méthode')
ax3.set_ylabel('Temps d\'exécution (s)')

# Mémoire utilisée
ax3.plot(methods, memory_used, color='green', marker='o', label='Mémoire (Mo)', linestyle='--')
ax3.set_ylabel('Mémoire utilisée (Mo)')

# Afficher les graphiques
plt.show()
```

Cependant, l'approche One-Shot reste efficace et pertinente, tandis que le Zero-Shot fournit un résumé de base utile mais moins complet.

# **LoRA: Low Rank Adaption**

Le fine-tuning LoRA permet de remédier aux limites du fine-tuning traditionnel.

L'idée principale est de décomposer les matrices de pondération (toutes ou certaines) du modèle d'origine en matrices de rang inférieur et de les entraîner à la place.

**Metrics:**

Résultats :
- Similarité Cosinus : Mesure la similitude entre les embeddings des deux textes.
- Scores ROUGE : Mesure la précision, le rappel et le F-score pour rouge1, rouge2, et rougeL.
- BLEU Score : Mesure la qualité du texte généré en fonction de la correspondance des n-grammes entre le texte généré et la référence.

```python
# @title
import psutil
import time

def measure_performance(model):
    # Mesurer le nombre de paramètres
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    all_params = sum(p.numel() for p in model.parameters())
    param_ratio = 100 * trainable_params / all_params

    # Mesurer la mémoire utilisée
    mem_usage = psutil.Process().memory_info().rss / (1024 ** 2)  # En Mo
    return trainable_params, all_params, param_ratio, mem_usage

def compute_cosine_similarity(reference, generated, tokenizer, model):
    # Tokenisation des deux textes (référence et généré)
    reference_encoding = tokenizer(reference, return_tensors="pt", padding=True, truncation=True, max_length=512)
    generated_encoding = tokenizer(generated, return_tensors="pt", padding=True, truncation=True, max_length=512)

    with torch.no_grad():
        # Utilisation de BartModel pour obtenir les embeddings
        model = BartModel.from_pretrained('facebook/bart-large')  # On utilise BartModel pour extraire les embeddings
        reference_embedding = model(**reference_encoding).last_hidden_state.mean(dim=1)
        generated_embedding = model(**generated_encoding).last_hidden_state.mean(dim=1)

    # Calcul de la similarité cosinus
    similarity = cosine_similarity(reference_embedding.cpu(), generated_embedding.cpu())
    return similarity[0][0]  # Retourner la valeur de similarité cosinus

# Fonction pour calculer les scores ROUGE
def compute_rouge(reference, generated):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, generated)
    return scores


def compute_bleu(reference, generated):
    reference_tokens = [reference.split()]  # BLEU attend des listes de tokens
    generated_tokens = generated.split()   # Découper le texte généré en tokens
    bleu_score = sentence_bleu(reference_tokens, generated_tokens)
    return bleu_score

```

# **RAG (Retrieval-Augmented Generation)**

Retrieval Augmented Generation (RAG) est une autre façon plutôt intéressante d'enrichir les réseaux neuronaux avec des informations supplémentaires, sans avoir à affiner le modèle.

Ceci est illustré ci-dessous :

### Explication simple des étapes de RAG

#### 1. Recherche de documents (Retrieval)
- Avant de générer une réponse, RAG commence par rechercher dans une base de données ou un ensemble de documents pour trouver des informations pertinentes.
- Cela se fait généralement par une recherche de passages en utilisant des modèles d'encodage (ex. BERT, DPR) pour comparer la question avec un grand corpus de textes.
- L'idée est d'identifier les informations les plus utiles qui peuvent aider à répondre à la question posée.

#### 2. Génération de texte (Generation)
- Une fois que les passages pertinents sont trouvés, RAG utilise un modèle de génération, tel que **BART** ou **T5**, pour créer une réponse ou un résumé basé sur ces informations extraites.
- La réponse générée prend en compte les passages récupérés dans la première étape, augmentant ainsi la qualité de la réponse par rapport à une réponse générée sans information externe.

### Processus détaillé de RAG

#### 1. Encodage des questions et des documents
- La question est encodée en un vecteur numérique (**embedding**) à l'aide d'un modèle comme **DPR (Dense Passage Retrieval)** ou **T5**. De même, les documents ou passages du corpus sont également encodés de manière similaire.

#### 2. Recherche dans la base de données (Retrieval)
- Le modèle compare l'embedding de la question avec les embeddings des passages dans une base de données ou un index (souvent construit avec **FAISS**, une bibliothèque de recherche vectorielle) pour trouver les passages les plus pertinents.

#### 3. Fusion des informations récupérées
- Les passages récupérés sont ensuite concaténés avec la question originale et envoyés à un modèle de génération de texte, tel que **BART**, pour créer une réponse.

#### 4. Génération de la réponse
- Le modèle de génération prend en compte à la fois la question et les passages pertinents et génère une réponse qui est plus précise et informative qu'une réponse générée uniquement à partir de la question.

### Avantages de RAG

- **Amélioration de la qualité des réponses** : En combinant la recherche d'information avec la génération, RAG peut produire des réponses qui sont plus ancrées dans des données réelles et récentes.
- **Accès à un large corpus de connaissances** : Le modèle n'a pas besoin de mémoriser toutes les informations durant son entraînement. Il peut récupérer des informations à la volée à partir d'une base de données externe.
- **Flexibilité** : RAG peut être appliqué à une large gamme de tâches, comme la réponse à des questions, la génération de résumés, etc.

```python
# @title
import time
import torch
import numpy as np
from transformers import BartForConditionalGeneration, BartTokenizer, DPRContextEncoder, DPRContextEncoderTokenizer, DPRQuestionEncoder, DPRQuestionEncoderTokenizer
from datasets import load_dataset
from rouge_score import rouge_scorer
from sklearn.metrics.pairwise import cosine_similarity
import faiss
import tracemalloc  # Pour la mesure de l'utilisation de la mémoire

# Charger les données et les modèles
dataset = load_dataset("cnn_dailymail", "3.0.0")
model_name = "facebook/bart-large-cnn"
model = BartForConditionalGeneration.from_pretrained(model_name)
tokenizer = BartTokenizer.from_pretrained(model_name)

# Modèles d'encodage pour le module de récupération (DPR encoders)
question_encoder = DPRQuestionEncoder.from_pretrained("facebook/dpr-question_encoder-single-nq-base")
context_encoder = DPRContextEncoder.from_pretrained("facebook/dpr-ctx_encoder-single-nq-base")
question_tokenizer = DPRQuestionEncoderTokenizer.from_pretrained("facebook/dpr-question_encoder-single-nq-base")
context_tokenizer = DPRContextEncoderTokenizer.from_pretrained("facebook/dpr-ctx_encoder-single-nq-base")

# Limiter le sous-ensemble à 10 articles
subset_size = 10
dataset_subset = dataset['train'].select(range(subset_size))

# Construire l'index FAISS pour stocker les embeddings des contextes
index = faiss.IndexFlatIP(768)
context_embeddings = []

for example in dataset_subset:
    ctx_input = context_tokenizer(example['article'], return_tensors="pt", max_length=512, truncation=True)
    ctx_embedding = context_encoder(ctx_input["input_ids"]).pooler_output.detach().numpy()
    index.add(ctx_embedding)
    context_embeddings.append(ctx_embedding)

# Fonction de récupération et génération de résumé
def retrieve_and_summarize(article):
    question_inputs = question_tokenizer(article, return_tensors="pt", max_length=512, truncation=True)
    question_embedding = question_encoder(question_inputs["input_ids"]).pooler_output.detach().numpy()

    # Recherche des passages les plus similaires
    D, I = index.search(question_embedding, k=3)
    retrieved_passages = [dataset_subset[int(i)]['article'] for i in I[0]]

    # Concaténer les passages récupérés avec l'article
    combined_input = " ".join(retrieved_passages) + " " + article
    inputs = tokenizer(combined_input, return_tensors="pt", max_length=1024, truncation=True)

    # Générer le résumé
    summary_ids = model.generate(inputs["input_ids"])
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary

# Initialiser les métriques
rouge_scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL', 'rougeLsum'], use_stemmer=True)
rouge_scores = {'rouge1': [], 'rouge2': [], 'rougeL': [], 'rougeLsum': []}
cosine_similarities = []
execution_times = []
memory_usages = []

# Évaluation des résumés
for example in dataset_subset:
    article = example["article"]

    # Mesurer l'utilisation de la mémoire avant l'exécution
    tracemalloc.start()
    start_time = time.time()

    # Récupérer et générer le résumé
    summary = retrieve_and_summarize(article)

    # Temps d'exécution et utilisation de la mémoire
    execution_times.append(time.time() - start_time)
    _, memory_peak = tracemalloc.get_traced_memory()
    memory_usages.append(memory_peak / (1024 * 1024))  # Convertir en Mo
    tracemalloc.stop()

    # Résumé de référence
    reference_summary = example["highlights"]

    # Calcul des scores ROUGE
    scores = rouge_scorer.score(reference_summary, summary)
    for key in rouge_scores:
        rouge_scores[key].append(scores[key].fmeasure)

    # Calcul de la similarité cosinus entre les embeddings des résumés
    summary_embedding = question_encoder(question_tokenizer(summary, return_tensors="pt")["input_ids"]).pooler_output.detach().numpy()
    reference_embedding = question_encoder(question_tokenizer(reference_summary, return_tensors="pt")["input_ids"]).pooler_output.detach().numpy()
    cosine_sim = cosine_similarity(summary_embedding, reference_embedding)
    cosine_similarities.append(cosine_sim[0][0])

# Calcul des moyennes des métriques
avg_execution_time = np.mean(execution_times)
avg_memory_usage = np.mean(memory_usages)
avg_rouge1 = np.mean(rouge_scores['rouge1'])
avg_rouge2 = np.mean(rouge_scores['rouge2'])
avg_rougeL = np.mean(rouge_scores['rougeL'])
avg_rougeLsum = np.mean(rouge_scores['rougeLsum'])
avg_cosine_similarity = np.mean(cosine_similarities)

# Affichage des résultats
print("Résultats de l'évaluation :")
print(f"Temps d'exécution moyen (s) : {avg_execution_time:.4f}")
print(f"Utilisation mémoire moyenne (Mo) : {avg_memory_usage:.2f}")
print(f"ROUGE-1 : {avg_rouge1:.4f}")
print(f"ROUGE-2 : {avg_rouge2:.4f}")
print(f"ROUGE-L : {avg_rougeL:.4f}")
print(f"ROUGE-Lsum : {avg_rougeLsum:.4f}")
print(f"Similarité Cosinus moyenne : {avg_cosine_similarity:.4f}")
```

# Introduction au Renforcement Apprentissage par Feedback Humain (RLHF)

**Le Renforcement Apprentissage par Feedback Humain** (RLHF) est une méthode d'apprentissage supervisée qui combine l'apprentissage par renforcement (RL) avec des retours humains pour améliorer les performances d'un modèle d'IA. En utilisant des retours humains, on aide le modèle à prendre des décisions qui sont plus alignées avec les préférences humaines et à améliorer la qualité des réponses dans des tâches complexes.

---

## Pourquoi RLHF ?

Les modèles de langage avancés, bien qu'extrêmement performants, peuvent générer des résultats indésirables ou incohérents en raison de la complexité des réponses attendues. RLHF aide à affiner le modèle en le "guidant" via des retours humains qui orientent son comportement vers des réponses plus pertinentes et cohérentes.

---

## Processus de RLHF : Étapes Principales

### 1. **Pré-entraînement du modèle**
   - **But** : Créer une base solide de compréhension linguistique.
   - **Méthode** : On utilise un grand ensemble de données non supervisé pour pré-entraîner le modèle de langage. Cela lui permet d’apprendre la syntaxe, la grammaire, et un certain niveau de compréhension contextuelle.
   - **Résultat** : Un modèle de langage général, mais non optimisé pour des tâches spécifiques ou des préférences humaines.

### 2. **Supervised Fine-Tuning (SFT)**
   - **But** : Ajuster le modèle pour qu’il réponde dans un style ou format souhaité.
   - **Méthode** : Les annotateurs humains fournissent des réponses exemples à des questions ou des prompts spécifiques. Le modèle est ensuite ajusté pour apprendre de ces réponses.
   - **Résultat** : Le modèle acquiert un comportement souhaité de base en fonction des réponses humaines fournies.

### 3. **Génération de Comparaisons et Classement**
   - **But** : Recueillir des préférences humaines directes.
   - **Méthode** : Les annotateurs examinent plusieurs réponses générées par le modèle pour le même prompt, puis les classent selon leur qualité ou préfèrent une réponse par rapport à une autre.
   - **Résultat** : Un ensemble de données de préférences qui indique comment les réponses se comparent entre elles selon des critères humains.

### 4. **Apprentissage par Renforcement (RL)**
   - **But** : Utiliser les préférences humaines pour ajuster les réponses.
   - **Méthode** : Un modèle de récompense est entraîné en fonction des comparaisons humaines (plus une réponse est préférée, plus elle est "récompensée"). Le modèle utilise ensuite un algorithme d’apprentissage par renforcement (comme PPO - Proximal Policy Optimization) pour maximiser les récompenses.
   - **Résultat** : Un modèle qui génère des réponses de plus en plus alignées avec les préférences humaines.

---

## Détails du Processus de RLHF

1. **Recueillir des Retours Humains** : Les humains comparent différentes sorties du modèle et indiquent leurs préférences. Ces comparaisons sont essentielles pour enseigner au modèle à comprendre la qualité d'une réponse selon des critères humains.
   
2. **Modèle de Récompense** : Ce modèle est utilisé pour évaluer et quantifier la qualité d'une réponse en fonction des retours humains. Il aide le modèle principal à comprendre quelle réponse est plus proche des attentes humaines.

3. **Optimisation avec RL** : Avec le modèle de récompense, un algorithme RL ajuste le modèle principal pour maximiser la satisfaction des retours humains, en améliorant la qualité des sorties dans des contextes similaires.

---

## Avantages du RLHF

1. **Alignement avec les Préférences Humaines** : En incluant les retours humains, RLHF permet aux modèles de générer des réponses plus cohérentes avec ce que les utilisateurs attendent.
   
2. **Amélioration Continue** : Le modèle peut être continuellement amélioré avec de nouveaux retours, le rendant adaptable aux évolutions des préférences ou besoins.

3. **Réduction des Biais Indésirables** : RLHF permet de réduire les biais indésirables en orientant le modèle vers des réponses plus nuancées et équilibrées.

4. **Performance Accrue pour des Tâches Complexes** : RLHF est particulièrement utile pour des tâches complexes ou à faible signal où les réponses optimales sont difficiles à définir sans intervention humaine.

---

## Limites et Défis de RLHF

- **Dépendance aux Annotateurs** : Le modèle dépend fortement des annotateurs humains, ce qui peut introduire des biais subjectifs.
- **Coût en Ressources** : Recueillir des retours humains et entraîner le modèle avec RL est coûteux en termes de temps et de calculs.
- **Qualité des Données de Comparaison** : La qualité du modèle de récompense dépend directement de la qualité des données de comparaison.

---

## Conclusion

RLHF est une méthode puissante pour améliorer les modèles de langage avancés, en les rendant plus adaptés aux attentes humaines. Grâce à un processus en plusieurs étapes, RLHF intègre des retours humains pour guider les réponses générées, tout en améliorant la précision et la pertinence du modèle dans des contextes variés. Cependant, sa mise en œuvre est coûteuse et nécessite une collecte de données de qualité pour assurer des performances optimales.

```python
import torch
from datasets import load_dataset
from transformers import BartForConditionalGeneration, BartTokenizer
from peft import get_peft_model, LoraConfig
from torch.optim import Adam
import evaluate
import matplotlib.pyplot as plt

# Charger le modèle BART et le tokenizer
bart_tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
bart_model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")

# Geler toutes les couches sauf les deux dernières
for name, param in bart_model.named_parameters():
    if not any(layer in name for layer in ["model.decoder.layers[-1]"]):
        param.requires_grad = False

# Charger le dataset CNN/DailyMail
dataset = load_dataset("cnn_dailymail", "3.0.0", split="train[:1%]")

# Charger la métrique ROUGE
rouge_metric = evaluate.load("rouge")

# Configurer LoRA pour un modèle BART
peft_config = LoraConfig(
    r=4,
    lora_alpha=16,
    lora_dropout=0.1,
)
bart_model = get_peft_model(bart_model, peft_config)

# Configurer l’optimiseur
optimizer = Adam(filter(lambda p: p.requires_grad, bart_model.parameters()), lr=1e-5)

# Boucle d'entraînement avec RLHF simplifiée
num_epochs = 1

for epoch in range(num_epochs):
    for i, sample in enumerate(dataset):
        text = sample['article']
        reference_summary = sample['highlights']

        # Génération d'un résumé
        inputs = bart_tokenizer(text, max_length=1024, truncation=True, return_tensors="pt")
        summary_ids = bart_model.generate(inputs["input_ids"], max_length=128, num_beams=4, length_penalty=2.0, early_stopping=True)
        generated_summary = bart_tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        # Calcul des scores ROUGE pour utiliser comme récompense
        rouge_scores = rouge_metric.compute(predictions=[generated_summary], references=[reference_summary])
        reward = rouge_scores['rouge1']

        # Calcul de la perte
        reward_tensor = torch.tensor(reward, requires_grad=True)
        loss = -reward_tensor

        # Rétropropagation et optimisation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if i % 100 == 0:
            print(f"Epoch {epoch+1}/{num_epochs}, Step {i}, Reward: {reward:.4f}")
        if i >= 300:
            break

# Sauvegarder le modèle entraîné
bart_model.save_pretrained("trained_bart_model")

print("Entraînement terminé.")
```

**Recommandation** : **Prompt Tuning** offre le meilleur compromis entre précision, temps et consommation de mémoire, en particulier si nous souhaitons un bon équilibre entre performance et efficacité des ressources.
